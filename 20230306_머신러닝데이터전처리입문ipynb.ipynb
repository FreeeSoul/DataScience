#.5 이제 데이터 세트 완성하자!
지금까지 object값들-> 수치값으로 바꾸거나, 원핫인코딩으로 더미로 만들어놨으니 이제 그 더미들을 합치는 작업도 할 것이다

```python
print(data_.dtypes) # 이것은 항목이 여러 개인 표니까 타입을 확인하고 싶으면 .dtypes를 해야 한다.
print(data_['road_rating'].dtype) # 반면 이것은 항목 하나에 대한 타입이 궁금한 것이니 .dtype 으로 따진다.
print(str(data_['road_rating'].dtype) == 'int64') # 만약 타입을 검사하고 싶은 것이라면, 먼저 str(data_['road_rating'].dtype 로 타입을 문자열로 했을 때 어떻게 나오나 확인한뒤 'int64'처럼 쓰면 된다.
```

```python
i_col_arr = []
col_arr = data_.columns
for co in col_arr:
  if str(data_[co].dtype) != 'object': ## 이 작업을 하는 이유 -> 교재 83페이지에 나오듯 먼저 수치값으로 이루어진 데이터들만 남겨서 tmp 표를 만들어야 하므로, 항목마다 값이 object인 것들은 빼고
  # 나머지만 표를 구성하도록 코드 짰다
    i_col_arr.append(co)

i_col_arr
```

```python
str(data_['id'].dtype)
```

```python
i_col_arr
```

```python
tmp1 = data_[i_col_arr]
tmp1.head()
```

#6. 이제 아까 object값을 가져서 원핫인코딩으로 변형시켜줬던 항목들 dummies도 합쳐주자

```python
tmp2 = pd.concat([tmp1, data_dummies], axis=1)
tmp2.head()
```

```python
data_dummies
```

```python
tmp3 = pd.concat([tmp2, data_nodeNameDummies], axis=1)
tmp4 = pd.concat([tmp3, data_nodeNameDummies], axis=1)
tmp5 = pd.concat([tmp4, data_end_node_name_dummies], axis=1)

tmp5.to_csv('finish.csv', index=False)
```
